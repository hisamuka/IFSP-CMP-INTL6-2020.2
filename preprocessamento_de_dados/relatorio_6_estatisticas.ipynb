### Qual é o bairro com o aluguel mais caro?

#### Valor mais caro

```python
maiores_alugueis_por_bairro = grupo_por_bairro['Valor'].max()
maiores_alugueis_por_bairro  # não retorna ordenado pelo valor
```

```python
maiores_alugueis_por_bairro.sort_values(ascending=False)  # ordena os maiores aluguéis por bairro em ordem decrescente
```

Note que o maior valor de aluguel do dataset é 4500000.0 de um imóvel do bairro _Botafogo_. <br/>
Provavelmente na hora de coletar/construir o dataset, alguém digitou o valor de venda do imóvel ao invés de seu aluguel. <br/>
Portanto, este registro de imóveis é um **ruído** no dataset, ou também chamado de **_outlier_**. Precisamos remover casos assim em nossa base, para que nossa análise fique mais _fidedígna_, sem que ruídos assim interfiram em nossa análise.

O mesmo problema parece acontecer para o imóvel do _Recreio dos Bandeirantes_.

#### Aluguel mais caro na média

```python
grupo_por_bairro['Valor'].mean().sort_values(ascending=False)
```

Os _outliers_ (bairros com valores de aluguel muito discrepantes) parecem não ter influenciado/prejudicado tanto a **média** dos imóveis. "Parece". Mas, vamos analisar seu **desvio padrão**.

```python
grupo_por_bairro['Valor'].std().sort_values(ascending=False)
```

Note como o desvio padrão dos bairros com _outliers_ são enormes, o que faz com que a média de seus imóveis **NÃO sejam confiáveis**.

Note a quantidade de pontos "fora da curva/distribuição", ou seja, imóveis com valor de aluguel muito fora do comum. <br>
Vamos então visualizar apenas alguns destes outliers:

```python
selecao = dataset_residencial['Valor'] >= 200000
dataset_residencial.loc[selecao]
```

#### Removendo os outliers

<img src="boxplot_with_outliers.png" width="70%"/>

```python
Q1 = dataset_residencial['Valor'].quantile(0.25)
Q3 = dataset_residencial['Valor'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - (1.5 * IQR)
limite_superior = Q3 + (1.5 * IQR)
```

```python
Q1, Q3, IQR, limite_inferior, limite_superior
```

```python
selecao = (dataset_residencial['Valor'] >= limite_inferior) & (dataset_residencial['Valor'] <= limite_superior)
dataset_residencial_sem_outliers = dataset_residencial.loc[selecao]
dataset_residencial_sem_outliers
```

```python
dataset_residencial.shape
```

```python
plt.figure(figsize=(20, 12))
dataset_residencial_sem_outliers.boxplot('Valor')
```

Note que o seaborn recomputa o boxplot já indicando possíveis _outliers_ para este novo dataset. <br/>
Porém, cabe a você interpretar se realmente estes são de fato _outliers_ ou não.

Aparentemente, parece que eles não são _outliers_, pois de fato há imóveis de luxo com aluguel > do que 10 mil reais, por exemplo. <br/>
Então, neste caso, vamos assumir que o novo dataset está correto.

A fim de ter certeza disso, você poderia analisar **outras variáveis** para esses casos de _outliers_. P. ex, se você identificar um imóvel com um aluguel muito alto, porém cujo bairro possui aluguéis muito mais baratos, é de se desconfiar que este imóvel é um _outlier_.

### Explorando mais os dados em busca de _outliers_

A **identificação de _outliers_** que fizemos no exemplo passado, utilizou os valores de aluguéis de **todos** os tipos de imóveis para tal. <br/>
Note que ao fazer isso, os valores _elevados_ de um dado tipo de imóvel (p. ex, apartamento) podem influenciar muito o cômputo dos _quartis_. <br/>
Desta forma, tipos de imóveis cujos aluguéis são consideravelmente _mais baixos_ (p. ex., quitinete), podem se tornar _outliers_ por juntar todos os tipos neste cômputo.

#### Oberservando a distribuição de aluguéis por tipo de imóvel

```python
# estamos utilizando o dataset original
dataset_residencial.head()
```

```python
plt.figure(figsize=(20, 12))

sns.boxplot(data=dataset_residencial, x='Tipo', y='Valor')
```

Apenas analisando os boxplots dos aluguéis para cada tipo de imóvel, é possível identificar que os casos com mais _outliers_ são imóveis do tipo 'Apartamento' e 'Casa de Condomínio'.

#### Identificando outliers por tipo de imóvel

```python
grupo_tipo_imovel = dataset_residencial.groupby('Tipo')
grupo_tipo_imovel
```

```python
grupo_tipo_imovel.groups
```

<img src="boxplot_with_outliers.png" width="70%"/>

```python
# computando os quartiles, IQR, e limites inferiores e superiores para cada tipo de imóvel, independentemente
Q1 = grupo_tipo_imovel['Valor'].quantile(0.25)
Q3 = grupo_tipo_imovel['Valor'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - (1.5 * IQR)
limite_superior = Q3 + (1.5 * IQR)
```

```python
Q1
```

```python
Q3
```

```python
IQR
```

```python
limite_inferior
```

```python
limite_superior
```

```python
limite_superior.loc['Apartamento']  # qual é o limite superior para o tipo 'Apartamento'
```

```python
### Removendo os outliers para cada tipo de imóvel especificamente, separadamente
tipos_de_imoveis = dataset_residencial['Tipo'].unique()

dataset_residencial_sem_outliers_por_tipo = pd.DataFrame()  # cria um DataFrame/Tabela vazia

for tipo in tipos_de_imoveis:
    limite_inferior_do_tipo = limite_inferior.loc[tipo]
    limite_superior_do_tipo = limite_superior.loc[tipo]
    
    eh_do_tipo = dataset_residencial['Tipo'] == tipo  # eh_do_tipo é uma Series, cuja qtde de elementos é igual ao número de linhas da tabela
    dataset_do_tipo = dataset_residencial.loc[eh_do_tipo]  # seleciona as linhas do tipo desejado (eh_do_tipo True)
    
    selecao = (dataset_do_tipo['Valor'] >= limite_inferior_do_tipo) & (dataset_do_tipo['Valor'] <= limite_superior_do_tipo)  # se True, não eh outlier
    
    dataset_do_tipo_sem_outliers = dataset_do_tipo.loc[selecao]
    dataset_residencial_sem_outliers_por_tipo = pd.concat([dataset_residencial_sem_outliers_por_tipo, dataset_do_tipo_sem_outliers])

dataset_residencial_sem_outliers_por_tipo
```

```python
plt.figure(figsize=(20, 12))
sns.boxplot(data=dataset_residencial_sem_outliers_por_tipo, x='Tipo', y='Valor')
```

#### Salvando o dataset preprocessado sem outliers

```python
dataset_residencial_sem_outliers_por_tipo.to_csv('datasets/aluguel_residencial_preprocessado_com_variaveis_sem_outliers.csv', sep=';', index=False)
```
