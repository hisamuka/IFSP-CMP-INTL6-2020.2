# Relatório - Criando Variáveis

```python
import pandas as pd
dataset_residencial = pd.read_csv('datasets/aluguel_residencial_preprocessado.csv', sep=';')
dataset_residencial
```

### Criando a variável 'ValorBruto' (aluguel + condomínio + IPTU)

```python
# ele cria uma coluna nova ('ValorBruto') e faz a soma das três outras colunas, linha a linha
dataset_residencial['ValorBruto'] = dataset_residencial['Valor'] + dataset_residencial['Condominio'] + dataset_residencial['IPTU']
```

```python
dataset_residencial
```

### Criando a variável 'ValorPorM2'

```python
dataset_residencial['ValorPorM2'] = dataset_residencial['Valor'] / dataset_residencial['Area']
```

```python
dataset_residencial
```

### Criado a variável 'ValorBrutoPorM2'

```python
dataset_residencial['ValorBrutoPorM2'] = dataset_residencial['ValorBruto'] / dataset_residencial['Area']
```

```python
dataset_residencial
```

### Agregando Tipos de Imóveis em 'Casa' e 'Apartamento'
Queremos criar uma nova coluna, chamada 'TipoAgregado', onde queremos agregar alguns tipos de imóveis:
- Todos os tipos de imóveis residenciais relacionados a casas terão o tipo agregado 'Casa''
- Os tipos 'Apartamento' e 'Quitinete' terão o tipo agregado 'Apartamento'

```python
dataset_residencial['Tipo'].unique()
```

```python
tipo_agregado_casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# def checa_se_eh_casa(tipo, tipo_agregado_casa):
#     if tipo in tipo_agregado_casa:
#         return 'Casa'
#     else:
#         return 'Apartamento'

dataset_residencial['Tipo']
```

```python
dataset_residencial['TipoAgregado'] = dataset_residencial['Tipo'].apply(lambda tipo: 'Casa' if tipo in tipo_agregado_casa else 'Apartamento')
dataset_residencial
```

### Deletando Colunas

```python
dataset_residencial['ColunaTeste'] = dataset_residencial['Valor'] * dataset_residencial['Area']
dataset_residencial
```

```python
del dataset_residencial['ColunaTeste']
dataset_residencial
```

### Salvando o Dataset com as novas variaveis

```python
dataset_residencial.to_csv('datasets/aluguel_residencial_preprocessado_com_variaveis.csv', sep = ';', index=False)
```
