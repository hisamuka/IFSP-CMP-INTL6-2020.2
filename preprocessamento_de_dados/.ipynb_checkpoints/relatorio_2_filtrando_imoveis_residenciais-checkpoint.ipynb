### Filtrando apenas Imóveis Residenciais

```python
list(tipo_de_imoveis['Tipo'])
```

```python
tipos_residenciais = ['Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila']
tipos_residenciais
```

```python
dataset.head()
```

```python
# retorna uma Series de booleans informando se uma respectiva linha da tabela possui um Tipo
# que está dentro de (isin) --- é um dos elementos --- da lista/conjunto definidos em tipos_residenciais
dataset['Tipo'].isin(tipos_residenciais)
```

```python
selecao = dataset['Tipo'].isin(tipos_residenciais)
dataset_residencial = dataset.loc[selecao]
dataset_residencial
```

Note que os índices, na tabela filtrada, não estão de 0 até o tamanho da tabela - 1. <br/>
Temos apenas 22580 linhas mas temos índices, p. ex, iguais a 32957, 32958.

```python
dataset_residencial.index = range(dataset_residencial.shape[0])
dataset_residencial
```

### Exportando o Dataset Filtrado

```python
# vamos salvar o dataset com apenas dados de imóveis residenciais
# vamos deixar o mesmo separador que tinha no dataset original
# exportando o dataset dessa maneira, o pandas criará uma coluna para representar os índices do DataFrame
dataset_residencial.to_csv('datasets/aluguel_residencial.csv', sep = ';')
```

```python
# note a coluna Unnamed: 0
pd.read_csv('datasets/aluguel_residencial.csv', sep=';')
```

```python
# para corrigir isso, podemos dizer que não queremos salvar os índices
dataset_residencial.to_csv('datasets/aluguel_residencial.csv', sep = ';', index=False)
```
